<div class="alert alert-block alert-success">Implementation to show K - Means Clustering.

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
dataset = make_blobs(n_samples=200,centers=5,n_features=2,cluster_std=1.5,random_state=500)
dataset
```

```python
points = dataset[0]
points
```

```python
kmeans = KMeans(n_clusters=5)
```

```python
kmeans.fit(points)
```

```python
plt.scatter(dataset[0][:,0], dataset[0][:,1])
```

```python
clusters = kmeans.cluster_centers_
```

```python
print(clusters)
```

```python
y_km = kmeans.fit_predict(points)
y_km
```

```python
plt.scatter(points[y_km == 0,0],points[y_km==0,1], s=50, color = 'red')
plt.scatter(points[y_km == 1,0],points[y_km==1,1], s=50, color = 'blue')
plt.scatter(points[y_km == 2,0],points[y_km==2,1], s=50, color = 'yellow')
plt.scatter(points[y_km == 3,0],points[y_km==3,1], s=50, color = 'cyan')
plt.scatter(points[y_km == 4,0],points[y_km==4,1], s=50, color = 'lime')


plt.scatter(clusters[0][0], clusters[0][1], marker='*',s=200, color='black')
plt.scatter(clusters[1][0], clusters[1][1], marker='*',s=200, color='black')
plt.scatter(clusters[2][0], clusters[2][1], marker='*',s=200, color='black')
plt.scatter(clusters[3][0], clusters[3][1], marker='*',s=200, color='black')
plt.scatter(clusters[4][0], clusters[4][1], marker='*',s=200, color='black')
```

<div class="alert alert-block alert-success">2. Develop basic code for Decision Tree Algorithm.

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\ishan\Desktop\PPT\salary.csv")
df
```

```python
inputs = df.drop('salary_more_than_100k', axis='columns')
target = df['salary_more_than_100k']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
company = LabelEncoder()
job = LabelEncoder()
degree = LabelEncoder()
```

```python
inputs['company_label'] = company.fit_transform(inputs['company'])
inputs['job_label'] = job.fit_transform(inputs['job'])
inputs['degree_label'] = degree.fit_transform(inputs['degree'])
inputs
```

```python
inputs_label= inputs.drop(['company','job','degree'], axis='columns')
inputs_label
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_label,target)
```

```python
model.score(inputs_label,target)
```

```python
prediction = model.predict([[2,2,1]])
print(prediction)

if prediction == 1:
    print("The person will have salary over 100k")
else:
    print("The person will not have salary over 100k")
```
